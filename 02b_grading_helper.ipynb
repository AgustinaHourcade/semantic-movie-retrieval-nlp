# Grading Helper
Run the cells below to easily read the movie descriptions (narrative content) for each query in your evaluation pool. This will make it much faster to assign relevance scores in the `qrels_to_grade_english.csv` file.

```python
import pandas as pd

# 1. Load the evaluation pool CSV
qrels = pd.read_csv('qrels_to_grade_english.csv')

# 2. Load the original dataset to fetch the descriptions
movies = pd.read_pickle('movies_with_embeddings.pkl')

# 3. Create a dictionary mapping titles to their narrative content for fast lookups
movie_desc_dict = dict(zip(movies['title'], movies['content_narrative']))

print("Data loaded! Run the next cells to view the descriptions.")
```

```python
def show_movies_for_query(query_str):
    print(f"\n{'='*100}")
    print(f"\033[1mQUERY: '{query_str}'\033[0m")
    print(f"{'='*100}\n")
    
    subset = qrels[qrels['query'] == query_str]
    
    if subset.empty:
        print("No movies found for this query in the CSV.")
        return
        
    for idx, row in subset.iterrows():
        title = row['movie_title']
        desc = movie_desc_dict.get(title, "Description not found in dataset.")
        
        print(f"\033[1m🎬 {title}\033[0m")
        print(f"📝 {desc}\n")
        print("-" * 100 + "\n")
```

```python
# List all the queries you are grading to use as reference:
unique_queries = qrels['query'].unique()
for i, q in enumerate(unique_queries):
    print(f"{i}: {q}")
```

```python
# Show descriptions for the first query in your list.
# Change the index (e.g. unique_queries[1]) or pass the query string directly to view others!
query_to_view = unique_queries[0]
show_movies_for_query(query_to_view)
```
